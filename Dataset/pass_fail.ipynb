```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_score = pd.read_csv('./DotSlash_Scores_2.csv')
df_attendance = pd.read_csv('./Student_Attendance.csv')

df_concat = pd.concat([df_score, df_attendance], axis=1)
df_concat
```

```python
import random
list = []
for index, row in df_concat.iterrows():
    result = ""
    if row['Score']  + row['Attendance'] < 90:
        result = "Fail"
    else:
        result = "Pass"

    if row['Score'] > 90:
        list.append(result)
        continue
    elif row['Score'] < 35:
        list.append(result)
        continue

    rand_no = random.randint(1, 100)
    if rand_no <= 15:
        #print("Flip")
        if result == "Pass":
            result = "Fail"
            #print("Flip1")
        else:
            result = "Pass"
            #print("Flip 2")
    

    #df_score['Result'] = result
    list.append(result)
```

```python
df_attendance['result'] = list
```
