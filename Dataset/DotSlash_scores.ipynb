```python
bias_teacher = {}

bias_teacher['AIWRSurabhi'] = 2
bias_teacher['AIWRNage'] = 1
bias_teacher['AIWRBhaskar'] = 1

bias_teacher['BCShruthi'] = 1

bias_teacher['DFGBhaskar'] = -1
bias_teacher['DFGNage'] = 1

bias_teacher['IPNPVenkatesh'] = 2
```

```python
bias_course = {}

bias_course['AIWR'] = 2
bias_course['BC'] = 2
bias_course['GP'] = -1
```
