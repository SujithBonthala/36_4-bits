```python
def fun(df,attr, fac):
  return df[attr].where(df['Faculty'] == fac).mean()
```

```python
def recommend(course,df_score_path,df_feedback_path):
    df_score = pd.read_csv(df_score_path)
    df_feedback = pd.read_csv(df_feedback_path)

    #df_concat = pd.concat([df_score, df_feedback], axis=1)
    df_concat = pd.merge(df_score, df_feedback, on=['Course','Faculty','SRN','Full Name'])

    df_concat = df_concat[df_concat['Course']==course]

    fac = df_concat['Faculty'].unique()

    dic = {}

    for name in fac:
        dic[name] = 0

    sum = 0
    for name in fac:
        sum+=fun(df_concat,'Score',name) + fun(df_concat,'Well versed with the subject',name) + fun(df_concat,'Explains concepts in an understandable way',name) + fun(df_concat,'Use of presentations',name) + fun(df_concat,'Solves doubts willingly',name) + fun(df_concat,'Provides support for students going above and beyond',name)
        dic[name] = sum
        sum = 0

    sorted_dict_desc = dict(sorted(dic.items(), key=lambda x: x[1], reverse=True))

    lst = [x for x in sorted_dict_desc.keys()]
    


    return lst

    
```

```python
recommend('AIWR','../Dataset/DotSlash_Scores_2.csv','../Dataset/DotSlash_Feedback.csv')
```
